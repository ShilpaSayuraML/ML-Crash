## Evaluate ML Algorithms with Resampling

- know how well your ML algorithm perform on unseen data. 
- use resampling methods to make estimates for new data. 

1. Train and Test Sets.
2. k-fold Cross Validation.
3.  Leave One Out Cross Validation.
4. Repeated Random Test-Train Splits.

```python
# Evaluate using a train and a test set

from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = read_csv(filename, names=names)
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
X
```

```python
Y
```

```python
test_size = 0.33
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size,random_state=seed)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
result = model.score(X_test, Y_test)
```

```python
print("Accuracy: ", result*100.0)
```

## K-fold Cross Validation
- We estimate the performance of a machine learning algorithm with less variance than a single train-test set split. 
- We split the dataset into k-parts (e.g. k = 5 or k = 10). 
- Each split of the data is called a fold.
- The algorithm is trained on k âˆ’ 1 folds
- One held back and tested on the held back fold.
- This is repeated so that each fold of the dataset is given a chance to be the held back test set. 
- After running cross validation we end up with k different performance scores 
- we can summarize it using a mean and a standard deviation.
- This result is a more reliable estimate of the performance of the algorithm on new data.


```python

# Evaluate using Cross Validation

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
num_folds = 10
seed = 7
kfold = KFold(n_splits=num_folds, random_state=seed, shuffle=True)
kfold
```

```python
model = LogisticRegression()
results = cross_val_score(model, X, Y, cv=kfold)
results
```

```python
print("Accuracy:", results.mean()*100.0, results.std()*100.0)
```

## Leave One Out Cross Validation

We can configure cross validation size of the fold is 1
This variation is called leave-one-out cross validation. 



```python
# Evaluate using Leave One Out Cross Validation
from pandas import read_csv
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = read_csv(filename, names=names)
array = dataframe.values
```

```python
X = array[:,0:8]
Y = array[:,8]
num_folds = 10
loocv = LeaveOneOut()
```

```python
model = LogisticRegression()
```

```python
results = cross_val_score(model, X, Y, cv=loocv)
print("Accuracy", results.mean()*100.0, results.std()*100.0)
```

## Repeated Random Test-Train Splits
- create a random split of the data 
- repeat the process of splitting and evaluation multiple times,


```python

# Evaluate using Shuffle Split Cross Validation
from pandas import read_csv
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = read_csv(filename, names=names)
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]

```

```python
n_splits = 10
test_size = 0.33

seed = 7
kfold = ShuffleSplit(n_splits=n_splits, test_size=test_size, random_state=seed)
model = LogisticRegression()
```

```python
results = cross_val_score(model, X, Y, cv=kfold)
print("Accuracy", results.mean()*100.0, results.std()*100.0)
```
