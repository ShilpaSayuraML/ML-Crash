# Basic Statistics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

sns.set()
%matplotlib inline
```

```python
df = pd.read_csv("bank.csv", delimiter=";")
df.head()
```

## Numerical Data

### Single Variable

https://www.w3schools.com/statistics/index.php

```python
print("Mean - {}".format(df["age"].mean()))
print("Median - {}".format(df["age"].median()))
print("Min - {}".format(df["age"].min()))
print("Max - {}".format(df["age"].max()))
```

```python
df.describe()
```

```python
data = norm.rvs(10.0, 1, size=5000, random_state=0)
sns.displot(data, kde=False)
```

```python
data.std()
```

```python
sns.displot(df["age"], kde=False) # Show only histogram
```

```python
sns.displot(df["age"], bins=50, kde=False)
```

```python
sns.boxplot(df["age"], orient="v", width=0.2)
```

### Multiple Variables

```python
sns.regplot(x=df["age"], y=df["balance"], fit_reg=False) # Don't fit a regression line
```
