```python
# neural network with keras
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# load the dataset
dataset = loadtxt('pima-indians-diabetes.data.csv', delimiter=',')
```

The variables<br>

**Input Variables (X):**<br>

1. Number of times pregnant
2. Plasma glucose concentration at 2 hours in an oral glucose tolerance test
3. Diastolic blood pressure (mm Hg)
4. Triceps skin fold thickness (mm)
5. 2-hour serum insulin (mu U/ml)
6. Body mass index (weight in kg/(height in m)^2)
7. Diabetes pedigree function
8. Age (years)

**Output Variables (y):**<br> 

1. Class variable (0 or 1)

```python
dataset[0:3] # The data will be stored in a 2D array [rows, columns].
```

You can split the array into two arrays <br>
select first eight columns from index 0 to index 7 via the slice 0:8. <br>
then select the output column (the 9th variable) via index 8.

```python
# split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
X[:3]
```

```python
y[:3]
```

** Models in Keras are defined as a sequence of layers.**<br>
- ensure the input layer has the correct number of input features.
- specify it when creating the first layer with the input_shape argument and setting it to (8,) for presenting the eight input variables as a vector.

- Fully connected layers are defined using the Dense class. 
- Specify the number of neurons in the layer as the first argument
- and the activation function using the activation argument.

- Sigmoid and Tanh activation functions were preferred for all layers. <br><br>
- The model expects rows of data with 8 variables
- The first hidden layer has 12 nodes and uses the relu activation function.

- The second hidden layer has 8 nodes and uses the relu activation function.
- The output layer has 1 node and uses the sigmoid activation function.

https://towardsdatascience.com/multi-layer-neural-networks-with-sigmoid-function-deep-learning-for-rookies-2-bf464f09eb7f

```python

# define the keras model
model = Sequential()
model.add(Dense(12, input_shape=(8,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

** Compile Keras Model**
- training a network means finding the best set of weights to map inputs to outputs in your dataset.
<br>
- use cross entropy as the loss argument.
- choose loss functions based on your problem
- 
https://machinelearningmastery.com/how-to-choose-loss-functions-when-training-deep-learning-neural-networks/

**Learn**<br>
- configure a model for mean squared error and variants for regression problems.
- configure a model for cross-entropy and hinge loss functions for binary classification.
- configure a model for cross-entropy and KL divergence loss functions for multi-class classification.

- Regression Loss Functions
1. Mean Squared Error Loss
2. Mean Squared Logarithmic Error Loss
3. Mean Absolute Error Loss

https://machinelearningmastery.com/loss-and-loss-functions-for-training-deep-learning-neural-networks/

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Fit your model
-Train your model with loaded data by calling the fit() function

- Training occurs over epochs
- each epoch is split into batches.

- **Epoch**: One pass through all of the rows in the training dataset
- **Batch**: One or more samples considered by the model within an epoch before weights are updated<br>
https://machinelearningmastery.com/difference-between-a-batch-and-an-epoch/

```python
# fit the keras model on the dataset
model.fit(X, y, epochs=150, batch_size=10)
```

## Evaluate Keras Model
- evaluate the performance of the network on the same dataset.
- use the evaluate() function and pass it the same input and output used to train the model.
- This will generate a prediction for each input and output pair and collect scores, including the average loss and any metrics you have configured, such as accuracy.

- The evaluate() function will return a list with two values. The first will be the loss of the model on the dataset, and the second will be the accuracy of the model on the dataset. 


```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: ', accuracy*100,",",  "Loss ", _)
```

```python
# Make Predictions

# make probability predictions with the model
predictions = model.predict(X)

# round predictions 
rounded = [round(x[0]) for x in predictions]
```
