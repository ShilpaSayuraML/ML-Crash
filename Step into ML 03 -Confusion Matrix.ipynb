## Machine Learning - Confusion Matrix

Confusion matrix is a table used in classification problems to assess errors in the model.
<br>
The rows represent the actual outcomes and while columns represent the predictions.
<br>
Confusion matrixes can be created by predictions made from a logistic regression.
<br>

https://towardsdatascience.com/understanding-confusion-matrix-a9ad42dcfd62

```python
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

actual = numpy.random.binomial(1,.9,size = 1000)
predicted = numpy.random.binomial(1,.9,size = 1000)

```

```python
actual
```

```python
predicted
```

```python
confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

The Confusion Matrix created has four different quadrants:<br>
<br>
True Negative (Top-Left Quadrant)<br>
False Positive (Top-Right Quadrant)<br>
False Negative (Bottom-Left Quadrant)<br>
True Positive (Bottom-Right Quadrant)<br><br>
**True:** values were accurately predicted,<br>**False:** wrong prediction.



**Created Metrics**<br>

Accuracy, Precision, Sensitivity (Recall), Specificity, and the F-score<br>
<br>



**Accuracy**<br>
Accuracy measures how often the model is correct.<br><br>

(True Positive + True Negative) / Total Predictions


```python
Accuracy = metrics.accuracy_score(actual, predicted)
print(Accuracy)
```

**Precision**<br>
Of the positives predicted, what percentage is truly positive?
<br><br>
True Positive / (True Positive + False Positive)
<br>
Precision does not evaluate the correctly predicted negative cases:

```python

Precision = metrics.precision_score(actual, predicted)
print(Precision)
```

**Sensitivity (Recall)**<br>
what percentage are predicted positive Of all the positive cases<br>

looks at true positives and false negatives (positives predicted as negative).<br><br>

True Positive / (True Positive + False Negative)<br>


```python
Sensitivity_recall = metrics.recall_score(actual, predicted)
print(Sensitivity_recall)
```

**Specificity**<br>
How well the model is at prediciting negative results?,br>
<br>
looks at it from the persepctive of negative results.
<br><br>
True Negative / (True Negative + False Positive)
<br><br>


```python
#Its opposit of Recall, we use the recall_score function, taking the opposite position label:
Specificity = metrics.recall_score(actual, predicted, pos_label=0)
print(Specificity)
```

**F-score**<br>
F-score is the "harmonic mean" of precision and sensitivity.<br><br>

It considers both false positive and false negative cases and is good for imbalanced datasets.<br><br>

2 * ((Precision * Sensitivity) / (Precision + Sensitivity))
<br><br>
This score does not take into consideration the True Negative values:

```python
F1_score = metrics.f1_score(actual, predicted)
print(F1_score)
```
